#### 5. 선형모델을 이용한 CCTV 수 예측


##### 선형모델
- 선형 함수 수식을 만들어서 예측
- y= wx + b


```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression() # 모델 객체 생성
```

```python
linear_model.fit(merge_result[['전체인구수']],merge_result['소계'])
```

```python
linear_model.predict([[1000000],[5000000]])
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

y = 0.00130916 * X(전체인구수0) + 645.0664965847219

#### 6. 결과 시각화

```python
# 모델의 예측 결과와 실제 구의 CCTV 설치 수 사이의 오차
cctv_pre=linear_model.predict(merge_result[['전체인구수']])
cctv_pre
```

```python
merge_result['CCTV오차']=merge_result['소계'] - cctv_pre
```

```python
merge_result['CCTV 절대값 오차']=np.abs(merge_result['소계'] - cctv_pre)
```

```python
merge_result
```

```python
# CCTV bar 차트 시각화
merge_result['CCTV오차'].sort_values().plot(kind='bar')

# 마이너스 해결
# import matplotlib
# matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 종합 정보를 보여주는 시각화
plt.figure(figsize=(15,10)) # 가로, 세로 크기 설정
plt.scatter(merge_result['전체인구수'],
           merge_result['소계'], s=300, c=merge_result['CCTV 절대값 오차'],
           cmap='Accent')

# CCTV오차가 큰 상위 10개 지역 이름을 시각화해보자
CCTV_error_top10 = merge_result[['CCTV 절대값 오차',
                                '전체인구수',
                                '소계']].sort_values(ascending=False, by='CCTV 절대값 오차').head(10)
for i in range(len(CCTV_error_top10)) :
    plt.text(CCTV_error_top10['전체인구수'].iloc[i],
            CCTV_error_top10['소계'].iloc[i],
            CCTV_error_top10.index[i], fontsize=15)
    
#선형 모델의 예측 경향 시각화 
input_x = np.linspace(0,800000,100) # 균등한 간격으로 range 내의 숫자를 추출
result_y = linear_model.predict(input_x.reshape(-1,1))# 추출한 입력 x를 예측
plt.plot(input_x,result_y,color='red',lw=4,label='LineaRegression 예측선')
plt.legend() # 범례

plt.colorbar()
plt.xlabel('각 구별 인구수')
plt.ylabel('각 구별 CCTV 수')
plt.title('서울시 각 구별 CCTV 현황')
plt.savefig('서울시 CCTV현황 분석결과.png')
plt.show()
```
