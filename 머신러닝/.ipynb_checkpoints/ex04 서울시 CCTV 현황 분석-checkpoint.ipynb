```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression() #   
```

```python
linear_model.fit(merge_result[['']],merge_result[''])
```

```python
linear_model.predict([[1000000],[5000000]])
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

y = 0.00130916 * X(0) + 645.0664965847219

#### 6.  

```python
#      CCTV    
cctv_pre=linear_model.predict(merge_result[['']])
cctv_pre
```

```python
merge_result['CCTV']=merge_result[''] - cctv_pre
```

```python
merge_result['CCTV  ']=np.abs(merge_result[''] - cctv_pre)
```

```python
merge_result
```

```python
# CCTV bar  
merge_result['CCTV'].sort_values().plot(kind='bar')

#  
# import matplotlib
# matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
#    
plt.figure(figsize=(15,10)) # ,   
plt.scatter(merge_result[''],
           merge_result[''], s=300, c=merge_result['CCTV  '],
           cmap='Accent')

# CCTV   10   
CCTV_error_top10 = merge_result[['CCTV  ',
                                '',
                                '']].sort_values(ascending=False, by='CCTV  ').head(10)
for i in range(len(CCTV_error_top10)) :
    plt.text(CCTV_error_top10[''].iloc[i],
            CCTV_error_top10[''].iloc[i],
            CCTV_error_top10.index[i], fontsize=15)
    
#     
input_x = np.linspace(0,800000,100) #   range   
result_y = linear_model.predict(input_x.reshape(-1,1))#   x 
plt.plot(input_x,result_y,color='red',lw=4,label='LineaRegression ')
plt.legend() # 

plt.colorbar()
plt.xlabel('  ')
plt.ylabel('  CCTV ')
plt.title('   CCTV ')
plt.savefig(' CCTV .png')
plt.show()
```
