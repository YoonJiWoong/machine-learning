```python
from flask import Flask, render_template, request
import cv2
import io
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model # 딥러닝 모델 로딩을 위한 함수
```

```python
model = load_model('./data/keras_model.h5')
```

```python
app = Flask(__name__)
@app.route('/')
def index():
    return render_template('upload2.html')


@app.route('/fileUpload', methods=['GET','POST'])
def fileUpload():
    if request.method == 'POST': # 요청이 POST 방식일때
        f = request.files['file'] # 요청 객체에서 file 이름을 가진 데이터를 꺼내오기
        in_memory_file = io.BytesIO() # byre 단위로 파일을 읽기 위한 객체
        f.save(in_memory_file) # byte단위 파일로 저장
        data = np.fromstring(in_memory_file.getvalue(), dtype=np.uint8) # numpy 타입으로 데이터 읽기        
        
        # 이미지가 color일 경우
        color_image_flag = 1
        img = cv2.imdecode(data,color_image_flag) # opencv 활용해서 이미지 데이터로 변경
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
     
        display(plt.imshow(img_rgb)) # 이미지 그리기
        plt.show() # 이미지 보이기
        
        display(np.array(img_rgb).shape)
        
        # 이미지 스케일링
        img_scated = (np.array(img_rgb,dtype=np.float32)/127.0)-1
        
        pre = model.predict_classes(img_scated.reshape(1,224,224,3))
        
    return '당신이 업로드한 사진은{}입니다'.format("목캔디" if pre[0]==0 else "귀여운찬욱")
        
        
app.run()
```
